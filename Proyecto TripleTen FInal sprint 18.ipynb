## **5.-Modelado**

```python
# Codificar las variables categ√≥ricas para evitar la multicolinealidad (elimina la primera categor√≠a de cada columna)
df_encoded = pd.get_dummies(df_complete_outer, columns=cat_cols, drop_first=True)
```

```python
# Dividir las variables objetivo
target = df_encoded['churn']
features = df_encoded.drop(columns='churn')
```

```python
# Divisi√≥n en conjunto de entrenamiento y validaci√≥n
features_train, features_valid, target_train, target_valid = train_test_split(
    features, target, test_size=0.30, random_state=42)
```

##### **Modelo de arbol de desicion:**

```python
# Utiliza los datos transformados para entrenar un √°rbol de decisi√≥n.

# Crear el modelo de √°rbol de decisi√≥n con control el sobreajuste ajustando hiperpar√°metros:
model = DecisionTreeClassifier(max_depth=5, min_samples_leaf=10, random_state=42)

# Entrenar el modelo
model.fit(features_train, target_train)

# Predecir en el conjunto de validaci√≥n
predicted_valid = model.predict(features_valid)

# Desicion Tree
print(f'üå≥‚úÖDesicion Tree Accuracy: {accuracy_score(target_valid, predicted_valid):.2f}')
print(f'üå≥‚úÖDesicion Tree Precision: {precision_score(target_valid, predicted_valid):.2f}')
print(f'üå≥‚úÖDesicion Tree Recall: {recall_score(target_valid, predicted_valid):.2f}')
print(f'üå≥‚úÖDesicion Tree F1-score: {f1_score(target_valid, predicted_valid):.2f}')
```
